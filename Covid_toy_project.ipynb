```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ct['cough'].value_counts().plot.pie(autopct="%1.1f%%")
plt.legend()
```

```python
ct['gender'].value_counts().plot.bar(figsize=(6,5),color='g')
plt.legend()
plt.xlabel("Gender")
plt.ylabel("Count")
```

```python
ct['city'].value_counts().plot.pie(figsize=(8,8),autopct="%1.1f%%")
plt.legend()
```

```python
print("\nDuplicate Rows:",ct.duplicated().sum())
```

```python
print("\nValue Ranges:")
print(ct.describe(include='all'))
```

```python
plt.hist(ct['age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='gender', data=ct, palette='Set2')
plt.title('Gender Distribution')
plt.show()
```

```python
sns.histplot(ct['fever'], bins=10, kde=True, color='orange')
plt.title('Fever Distribution')
plt.xlabel('Fever')
plt.show()
```

```python
sns.countplot(x='city', hue='has_covid', data=ct, palette='coolwarm')
plt.title('COVID Status by City')
plt.show()
```

```python
sns.countplot(x='gender', hue='has_covid', data=ct, palette='Set1')
plt.title('COVID Status by Gender')
plt.show()
```

```python
sns.countplot(x='cough', hue='has_covid', data=ct, palette='Set3')
plt.title('Cough Severity vs COVID Status')
plt.show()
```

```python
sns.boxplot(x='has_covid', y='age', data=ct)
plt.title('Age vs COVID Status (Box Plot)')
plt.show()
```

```python
covid_counts = ct['has_covid'].value_counts()
covid_counts.plot(kind='bar')
plt.title('Total COVID Cases')
plt.xlabel('Has COVID')
plt.ylabel('Number of People')
plt.show()
```

```python
sorted_ct = ct.sort_values(by='age')
plt.plot(sorted_ct['age'], sorted_ct['fever'], marker='o', linestyle='-', color='purple')
plt.title('Fever by Age (Line Plot)')
plt.xlabel('Age')
plt.ylabel('Fever')
plt.show()
```

```python
plt.scatter(ct['age'], ct['fever'], color='darkblue')
plt.title('Scatter Plot: Age vs Fever')
plt.xlabel('Age')
plt.ylabel('Fever')
plt.show()
```

```python
city_counts = ct['city'].value_counts()
city_counts.plot(kind='barh', color='skyblue')
plt.title('Number of People per City')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

```python
ct['fever'].plot(kind='line', figsize=(8, 4), title='fever')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
city_table = pd.crosstab(ct['city'], ct['has_covid'])
print("COVID Frequency Table by City:")
print(city_table)
```

```python
numeric_ct = ct[['age', 'fever']]
sns.heatmap(numeric_ct.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
